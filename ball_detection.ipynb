```python
import cv2
import numpy as np
import time

cap = cv2.VideoCapture('ball.mp4')


start = time.time()
cY_start =0


if(cap.isOpened() == False):
    print("Error opening video file")




while(cap.isOpened()):
    ret, frame = cap.read()
    
    if ret ==  True:
        
        gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        
        kernel = np.ones((30, 30), np.uint8)
        
        closing = cv2.morphologyEx(gray, cv2.MORPH_OPEN, kernel)
        
        black_img = np.zeros((closing.shape[0],closing.shape[1],3))

        canny_edge = cv2.Canny(closing, 30, 180)
        
        contours, hierarchy = cv2.findContours(canny_edge, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
        for cnt in contours:
            area=cv2.contourArea(cnt)
           
            if area>100:
               
                M = cv2.moments(cnt)
                stop= time.time()
                dur = stop-start
                 
                print(dur)
                cX = int(M["m10"] / M["m00"])
                cY = int(M["m01"] / M["m00"])
                diff = cY-cY_start
                speed= diff/dur
                cY_start = cY
                start=stop
                
                print(diff)
                print(speed)
                if speed<0:
                    cv2.arrowedLine(black_img, ((cX-20),cY), (cX-20,cY-20),(255,255,255), 4,8,0,0.3)
                if speed>0:
                    cv2.arrowedLine(black_img, (cX-20,cY-20),((cX-20),cY) ,(255,255,255), 4,8,0,0.3)
                if speed==0:
                    continue
                    
                cv2.circle(black_img, (cX,cY), 5, (255,255,255), -1)
                cv2.putText(black_img,"("+str(cX)+","+str(cY)+")",(cX+6,cY),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,255,255),2)
                cv2.putText(black_img,str(speed)+"pixels/sec",(cX,(cY-15)),cv2.FONT_HERSHEY_SIMPLEX,0.5,(255,255,255),2)
                cv2.drawContours(canny_edge, contours, -1,(225,225,225), 3)
               
       
        
                
                 
        cv2.imshow('frame',frame)
    
       
        
        
        cv2.imshow("black",black_img)
    
        
 

    
        if cv2.waitKey(20)== ord('q'):
            break
    else:
        break
        

cap.release()
 
# Closes all the frames
cv2.destroyAllWindows()
```
